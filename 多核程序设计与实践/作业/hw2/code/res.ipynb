```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 画第1个图：折线图
x=[8, 2048, 4096, 10240, 16384]
y0 = [0.000193,0.047312,0.157847,0.576712,1.279435]
y1 = [0.000160, 0.058350, 0.216828, 1.051273, 2.657742]
y2 = [0.000164, 0.039906, 0.14313, 0.508098, 1.059883]
y3 = [0.000161, 0.035979, 0.142145, 0.458670, 0.975408]
y4 = [0.630909, 23.623844, 81.882156, 504.479889, 1312.428223]
plt.subplot(221)
plt.plot(x,y0)
# 画第2个图：散点图
plt.subplot(222)
plt.plot(x,y1)
# 画第3个图：饼图
plt.subplot(223)
plt.plot(x,y2)
# 画第4个图：条形图
plt.subplot(224)
plt.plot(x,y3)
plt.show()
```

```python
plt.plot(x,y0,label='Baseline')
plt.plot(x,y0,'o')
plt.plot(x,y1, label = 'V1')
plt.plot(x,y1,'o')
plt.plot(x,y2, label = 'V2')
plt.plot(x,y2,'o')
plt.plot(x,y3, label = 'V3')
plt.plot(x,y3,'o')
plt.xlabel('Matrix Size')
plt.ylabel('Running Time(s)')
plt.legend()
plt.savefig('Cuda.jpg')
plt.show()
```

```python
plt.plot(x,y1,label = 'Cuda')
plt.plot(x,y1,'o')
plt.plot(x,y4,label = 'OpenMP')
plt.plot(x,y4,'o')
plt.xlabel('Matrix Size')
plt.ylabel('Running Time(s)')
plt.legend()
plt.savefig('Cuda-OMP.jpg')
plt.show()
```
