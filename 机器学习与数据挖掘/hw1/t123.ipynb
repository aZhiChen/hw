```python
import random
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
#n is the total number
#Monte Carlo
def count_pi(n):
    times = 0
    pi = []
    while times < 100:
        i = 0
        count = 0
        while i < n:
            #产生一个0-1的随机数
            x = random.random() 
            y = random.random()
            if (pow(x,2) + pow(y,2)) < 1:
                count += 1
            i += 1
        pi.append(4 * (count / n))
        times += 1
    pi = np.array(pi)
    mean = np.mean(pi)
    var = np.var(pi)
    return mean, var
```

```python
N = [20, 50, 100, 200, 300, 500, 1000, 5000]
MEAN = []
VAR = []
print(type(VAR))
for n in N:
    mean, var = count_pi(n)
    MEAN.append(mean)
    VAR.append(var)
```

```python
print("maen is")
print(MEAN)
print("Var is ")
print(VAR)
```

```python
#蒙特卡洛求积分[0,1]x^3
def cal_x_3(n):
    times = 0
    res = []
    while times < 100:
        sum = 0
        i = 0
        while i < n:
            sum += pow(random.random(), 3)
            i += 1
        sum /= n
        res.append(sum)
        times += 1
    res = np.array(res)
    mean = np.mean(res)
    var = np.var(res)
    return mean, var
```

```python
#蒙特卡洛求积分[0,1]x^3
def cal_X_3(n):
    times = 0
    res = []
    while times < 100:
        sum = 0
        i = 0
        j = 0
        while i < n:
            y = random.random()
            if (y < pow(random.random(), 3)):
                #sum += y
                j += 1
            i += 1
        sum = j/n
        res.append(sum)
        times += 1
    res = np.array(res)
    mean = np.mean(res)
    var = np.var(res)
    return mean, var, res
```

```python
N = [5,10,20,30,40,50,60,70,80,100]
MEAN = []
VAR = []
for n in N: 
    mean, var, res = cal_X_3(n)
    MEAN.append(mean)
    VAR.append(var)
```
